```python
#Importación de librería pandas.
import pandas as pd
```

```python
#Lectura del csv
df = pd.read_csv('C:\\Users\\Federico\\Downloads\\Hackaton\\Project 1\\temperaturesbcndesde1780_2023_long.csv')
df

#Output
#Temperaturas medias por mes y por año.
#2928 rows × 4 columns
```

```python
#Filtrado entre 1900 y 2023
df_filtro_1900_2023 = df[(df['Any'] >= 1900) & (df['Any'] <= 2023)]

#Reseteo del índex
df_filtro_1900_2023.reset_index(drop=True, inplace=True)

#Eliminación de columna "Desc_Mes" -> No aportaba información relevante
df_filtro_1900_2023 = df_filtro_1900_2023[['Any', 'Mes', 'Temperatura']]

#Traducción al inlgés por pedido de backend ✅
df_filtro_1900_2023.rename(columns={'Any': 'Year'}, inplace=True)
df_filtro_1900_2023.rename(columns={'Temperatura': 'Temperature'}, inplace=True)
df_filtro_1900_2023.rename(columns={'Mes': 'Month'}, inplace=True)

#Visualización del df final
df_filtro_1900_2023
```

```python
#Exportación en formato .json para generar el endpoint ✅
df_filtro_1900_2023.to_json('Eng_Temp_1900_2023_Year_Month_Temperature.json', orient='records', lines=True)
```

```python
#Equipo de frontend solicita una porción del df comprendido entre los años 1950 hasta 2023 para hacer las primeras pruebas de visualización antes que backend tenga listo los endpoints:

#Filtrado 1950 a 2023
df_filtro_1950_2023 = df[(df['Any'] >= 1950) & (df['Any'] <= 2023)]

#Reseteo el índex
df_filtro_1950_2023.reset_index(drop=True, inplace=True)

#Eliminación de columna "Desc_Mes" -> No aportaba información relevante
df_filtro_1950_2023 = df_filtro_1950_2023[['Any', 'Mes', 'Temperatura']]

#Unificación en catalán para mantener un mismo idioma
df_filtro_1950_2023.rename(columns={'Temperatura': 'Temperature'}, inplace=True)

#Visualización del df final
df_filtro_1950_2023
```

```python
#Exportación en formato .json para que front pueda comenzar a graficar.
df_filtro_1950_2023.to_json('temperatures_1950_2023_Any_Mes_Temperature.json', orient='records', lines=True)
```
