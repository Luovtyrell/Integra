```python
import pandas as pd

df = pd.read_csv('C:\\Users\\Federico\\Downloads\\Hackaton\\Project 1\\precipitacionsbcndesde1786_2023_long.csv')
df

#Temperaturas medias por mes y por año.
#238 rows × 13 columns
```

```python
# Filtrar el DataFrame
df_filtro_1900_2023 = df[(df['Any'] >= 1900) & (df['Any'] <= 2023)]

# Resetear el índice
df_filtro_1900_2023.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
df_filtro_1900_2023 = df_filtro_1900_2023[['Any', 'Mes', 'Precipitacions']]

# Renombrar la columna "Any" a "Year"
df_filtro_1900_2023.rename(columns={'Any': 'Year'}, inplace=True)

# Renombrar la columna "Precipitacions" a "Precipitation"
df_filtro_1900_2023.rename(columns={'Precipitacions': 'Precipitation'}, inplace=True)

# Renombrar la columna "Mes" a "Month"
df_filtro_1900_2023.rename(columns={'Mes': 'Month'}, inplace=True)

# Exportar a .json
df_filtro_1900_2023.to_json('Eng_Precip_1900_2023_Year_Month_Precipitation.json', orient='records', lines=True)

df_filtro_1900_2023
```

```python
# Filtrar el DataFrame
df_filtro_1950_2023 = df[(df['Any'] >= 1950) & (df['Any'] <= 2023)]

# Resetear el índice
df_filtro_1950_2023.reset_index(drop=True, inplace=True)

df_filtro_1950_2023

# Mostrar el DataFrame resultante
df_filtro_1950_2023 = df_filtro_1900_2023[['Any', 'Mes', 'Precipitacions']]

df_filtro_1950_2023

df_filtro_1950_2023.to_json('Precipitacions_1950_2023_Any_Mes_Precipitacions.json', orient='records', lines=True)
```

```python
#Media de precipitación por año
# Agrupar por 'Any' (año) y calcular la media de 'Precipitacions'
AVG_precipitacions_1950_2023 = round(df_filtro_1950_2023.groupby('Any')['Precipitacions'].mean(),0)
AVG_precipitacions_1950_2023
```

```python
# Añadir columna con la media de precipitaciones para cada año
df_filtro_1950_2023['AVG Any'] = df_filtro_1950_2023.groupby('Any')['Precipitacions'].transform('mean').round(0)

# Mostrar el DataFrame con la nueva columna
df_filtro_1950_2023
```
